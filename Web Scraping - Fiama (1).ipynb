```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.fiama.in/best-seller&sort=sort_order&order=desc&page=1'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.content,'html.parser')
soup
```

# ACCESSING PRODUCT IMAGES

```python
ProductImage = soup.findAll('img',{'class':'swap-image'})

img_urls = []
for img in ProductImage:
    img_url = img.get('data-src')
    if img_url:
        img_urls.append(img_url)

img_urls
```

# ACCESSING PRODUCT NAMES

```python
ProductName = soup.findAll('h4',{'class':'product-name'})
ProductName
```

```python
Product_Name = []
for i in ProductName:
    Product_Name.append(i.get_text().replace("\n","").strip().split(",")[0])
Product_Name
```

# ACCESSING QUANTITIES

```python
import re

def remove_after_g_ml(item):
    return re.sub(r'(g|ml).*', r'\1', item)

Quantity = []
for i in ProductName:
    qa = i.get_text().replace("\n","").split(",")[1].strip().replace(" ","")
    cleaned_items = remove_after_g_ml(qa)
    Quantity.append(cleaned_items)
Quantity
```

# ACCESSING RATINGS

```python
caption_section = soup.findAll('div',{'class':'caption'})
caption_section

star_ratings = []
for r in caption_section:
    rating = r.find('span',{'class':'rating-star'})
    if rating:
        star_ratings.append(rating.text)
    else:
        star_ratings.append("NA")
        
star_ratings
```

# ACCESSING NUMBER OF PEOPLE WHO RATED

```python
caption_section = soup.findAll('div',{'class':'caption'})
#caption_section
No_of_Ratings = []
for r in caption_section:
    rating = r.find('div',{'class':'rating-wrp'})
    if rating:
        ra = rating.text
        No_of_Ratings.append(ra)
    else:
        No_of_Ratings.append("NA")

No_of_Ratings
```

```python
caption_section = soup.findAll('div',{'class':'caption'})
#caption_section
No_of_Ratings = []
for r in caption_section:
    rating = r.find('div',{'class':'rating-wrp'})
    if rating:
        ra = rating.text.split(".")[1].replace("Ratings","")
        No_of_Ratings.append(ra)
    else:
        No_of_Ratings.append("NA")

No_of_Ratings
```

```python
caption_section = soup.findAll('div',{'class':'caption'})

def remove_first_digit(s):
    return s[1:] if s.isdigit() else s


No_of_Ratings = []
for r in caption_section:
    rating = r.find('div',{'class':'rating-wrp'})
    if rating:
        ra = rating.text.split(".")[1].replace("Ratings","").strip()
        cleaned_data = remove_first_digit(ra)
        No_of_Ratings.append(cleaned_data)
    else:
        No_of_Ratings.append("NA")
        
No_of_Ratings


```

# ACCESSING OLD PRICE

```python
OldPrice = soup.findAll('span',{'class':'price-old'})
OldPrice
```

```python
Old_Price = []
for o in OldPrice:
    Old_Price.append(o.getText().strip())
Old_Price
```

# ACCESSING DISCOUNT

```python
Offer = soup.findAll('span',{'class':'sale sale-percentage'})
Offer
```

```python
Discount = []
for o in Offer:
    Discount.append(o.getText().replace("Off","").strip())
Discount
```

# ACCESSING NEW PRICE

```python
NewPrice = soup.findAll('span',{'class':'price-new'})
NewPrice
```

```python
New_Price = []
for n in NewPrice:
    New_Price.append(n.getText().strip())
New_Price
```

# CREATING DATAFRAME

```python
import pandas as pd
Fiama_Products = pd.DataFrame({
    "Fiama_Product_Image":img_urls,
    "Fiama Product Name":Product_Name,
    "Quantity":Quantity,
    "Rating":star_ratings,
    "No of Ratings":No_of_Ratings,
    "Old Price(Rs)": Old_Price,
    "Discount Offer":Discount,
    "New Price(Rs)":New_Price,
    
})

Fiama_Products
```

# CONVERTING AND STORING DATA FRAME IN CSV FILE 

```python
Fiama_Products.to_csv("Fiama_Products_Web_Scraping_Final.csv",index=False)
```
